```python
from binance.client import Client
import matplotlib.pyplot as plt
import pandas as pd
import mplfinance as fplt
import matplotlib.dates as mpl_dates
import pandas as pd
```

```python
candles = client.get_klines(
    symbol='ETHUSDT', interval=Client.KLINE_INTERVAL_1MINUTE)
```

```python
candledf = pd.read_csv('./ETHUSDT_2_weeks_ago_UTC_1_hour_ago_UTC')
```

```python

candledf.head()

```

```python
index  =  pd.to_datetime(candledf['Opentime'])
```

```python
candledf = candledf.set_index(pd.DatetimeIndex(index))
```

```python
candledf.index
```

```python
candledf = candledf.drop('Opentime', axis=1)
```

```python
candledf['Open'] = candledf['Open'].astype(float)
candledf['Close'] = candledf['Close'].astype(float)
candledf['High'] = candledf['High'].astype(float)
candledf['Low'] = candledf['Low'].astype(float)
candledf['Volume'] = candledf['Volume'].astype(float)
```

```python
fplt.plot(candledf, type='candle', style='charles',
            title='ETHUSDT',
            ylabel='Price ($)',
            ylabel_lower='Shares \nTraded',
            volume=True, 
            mav=(3,10,100), 
            )
```
